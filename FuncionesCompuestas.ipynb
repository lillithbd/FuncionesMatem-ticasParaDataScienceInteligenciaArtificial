### Librerías

```python
import matplotlib.pyplot as plt   # librería para graficar
import numpy as np                # librería para manejo de vectores y utilidades matemáticas
```

```python
N = 1000

x = np.linspace(-10,10, num=N)

def g(x):
  return x**2

#Lo que salga de g(x) lo vamos a introducir a otra nueva funcion

def f(x):
  return np.sin(x)

y = g(x)

#La salida de una función como argumento de otra
f_o_g = f(y)
f_o_g = f(g(x))


plt.grid()
plt.plot(x,f_o_g)
```
