```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000
x = np.linspace(-5,5, num=N)
```

## Función lineal
$$y=mx+b$$

```python
#mantener valores a lo largo de un proceso, predecir venta
#Regresa el mismo valor de entrada
def f(x):
  return x

plt.plot(x, f(x))
plt.grid()
```

## Función escalón o de Heaviside
$$H(x) = 
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

```python
#Sirve para hacer clasificaciones categóricos (si existe o no existe, valores binarios)
def H(x):
  Y = np.zeros(len(x))
  for idx,x in enumerate(x):
    if x>=0:
      Y[idx]=1
  return Y
    

N=1000


y = H(x)

plt.plot(x,y)
plt.grid()
```

## Función sigmoide

$$f(x)=\frac{1}{1-e^{-x}}$$

```python
#Sirve para la regresión logística
#Sirve para las probabilidades, siempre regresará valores entre 0 y 1
#Cuando los valores sean extremandamente grandes retornará 0
def f(x):
  return 1/(1 + np.exp(-x))
    

N=1000

y = f(x)

plt.plot(x,y)
plt.grid()
```

## Función tangente hiperbólica

$$f(x)=\frac{2}{1+e^{-2x}}-1$$

```python
#Funcion de escalamiento
#Va a tener valores entre -1 y 1
def f(x):
  return np.tanh(x)
    

N=1000

y = f(x)

plt.plot(x,y)
plt.grid()
```

## Función ReLU

$$R(x)=max(0,x)$$

```python
#Cuando todos los valores sean mayores a 0 comienza a crecer como una línea recta
#Para valores menores que 0 los vuelve 0
#Sirve para hacer simulación de neuronas muertas
def f(x):
  return np.maximum(x,0)
    

N=1000

y = f(x)

plt.plot(x,y)
plt.grid()
```
